### WAYS TO OPERATE WITH NULL VALUES

"isnull()" : TO GENERATE A BOOLEAN DATA INDICATING MISSING VALUES. 

"notnull()": OPPOSITE OF "isnull()". 

"dropna()" : RETURN A FILTERED DATA ("DROPING ALL NULL VALUES"). 

"fillna()" : FILLING THE DATA WITH MISSING VALUES.

```python
# FIND THE  NULL DATA VALUES IN DATAFRAME

# DATA  WILL  BE REPRESENTED IN BOOLEAN FORMATE WHETHER DATA IS ("TRUE" OR "FALSE").

# TRUE INDICATES THE NULL VALUE.

DATA.isnull()
```

```python
# FINDING TOTAL SUM() NULL VALUES.

DATA.isnull().sum()
```

## MEAN VALUES

```python
# THE "MEAN VALUE" INDICATES AVERAGE VALUE.

#TO FIND THE MEAN DATA, ADD  TOGETHER ALL OF YOUR  VALUES AND DIVIDE BY THE NUMBER OF ADDENDS. 

#FINDING THE MEAN VALUES

DATA_mean = DATA.mean()
```

```python
# FINDING THE MEAN VALUES 

DATA_mean
```

```python
# FILLING THE NULL VALUES WITH THE MEAN 

DATA.fillna(DATA_mean,inplace= True)
```

```python
# AFTER FILLING NULL VALUES

DATA
```

```python
#  SO ALL  THE NULL NULL VALUES  ARE REPLACED WITH  MEAN VALUES 
```

```python
DATA.isnull().sum()
```
